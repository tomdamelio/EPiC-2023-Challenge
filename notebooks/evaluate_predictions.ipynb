```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from helpers_scenario3 import *
```

```python
path = '../results/test/scenario_1/annotations/scenario_1'

files = os.listdir(path)
```

```python
def plot_valence_arousal(df, idx):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6), sharey=True)

    ax1.plot(df['time'], df['valence_pred'], label='Valence Predicted', linewidth=2, )
    ax1.plot(df['time'], df['valence'], label='Valence Original', linewidth=2, alpha = 0.5)
    
    ax1.set_title(f'Time Series {idx}: Valence')
    ax1.set_xlabel('Time')
    ax1.set_ylabel('Value')
    ax1.legend()

    ax2.plot(df['time'], df['arousal_pred'], label='Arousal Predicted', linewidth=2, )
    ax2.plot(df['time'], df['arousal'], label='Arousal Original', linewidth=2,alpha = 0.5)
    ax2.set_title(f'Time Series {idx}: Arousal')
    ax2.set_xlabel('Time')
    ax2.legend()

    plt.show()
```

```python
for i, file in enumerate(files[:20]):
    df = pd.read_csv(os.path.join(path, file))
    plot_valence_arousal(df, i+1)
```
